# Project: Alcohol Consumption Analysis

This repository contains the Python code and analysis performed on the `consumption.csv` dataset. The project aims to provide a comprehensive understanding of alcohol consumption patterns across various bars, alcohol types, and brands.

## Project Objectives:

* **Data Loading & Initial Exploration**: Load the `consumption.csv` dataset and perform initial checks for structure, data types, and missing values.
* **Data Preprocessing**: Convert date-time columns to appropriate formats and extract relevant time-based features.
* **Exploratory Data Analysis (EDA)**:
    * Summarize key statistics for numerical data.
    * Identify unique categories within categorical columns.
* **Consumption Pattern Analysis**:
    * Analyze daily consumption trends.
    * Determine total consumption by alcohol type and brand.
    * Assess consumption performance across different bars.
    * Identify peak consumption periods throughout the day.
* **Data Visualization**: Create insightful plots to visually represent consumption trends and patterns.
* **Key Insights & Summary**: Conclude with a summary of major findings and actionable insights derived from the analysis.

## Dataset: `consumption.csv`

The dataset contains detailed records related to alcohol consumption, including:
* `Date Time Served`: Timestamp of the consumption record.
* `Bar Name`: Name of the bar where the consumption occurred.
* `Alcohol Type`: Type of alcohol consumed (e.g., Rum, Wine, Vodka).
* `Brand Name`: Specific brand of alcohol consumed.
* `Opening Balance (ml)`: Opening stock of the alcohol in milliliters.
* `Purchase (ml)`: Amount of new purchase in milliliters.
* `Consumed (ml)`: Amount of alcohol consumed in milliliters.
* `Closing Balance (ml)`: Remaining stock of the alcohol in milliliters.

This project utilizes popular Python libraries such as `pandas` for data manipulation, `matplotlib.pyplot` for plotting, and `seaborn` for enhanced visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the CSV file into a DataFrame
df = pd.read_csv('consumption.csv')
```

```python
# Step 2: Display the first 5 rows to verify the data
print("First 5 rows of the DataFrame:")
print(df.head())
```

```python
# Step 3: Display column names and their data types
print("\nColumn names and their data types:")
print(df.info())
```

```python
# Step 4: Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Step 5: Generate descriptive statistics for numerical columns
print("\nDescriptive statistics for numerical columns:")
print(df.describe())
```

```python
# Step 6: Check unique values for categorical columns
print("\nUnique values in 'Bar Name':")
print(df['Bar Name'].unique())
```

```python
print("\nUnique values in 'Alcohol Type':")
print(df['Alcohol Type'].unique())
```

```python
print("\nUnique values in 'Brand Name':")
print(df['Brand Name'].unique())
```

```python
# Step 7: Convert 'Date Time Served' to datetime objects
df['Date Time Served'] = pd.to_datetime(df['Date Time Served'])
print("\n'Date Time Served' column converted to datetime objects.")
print(df.info())
```

```python
# Step 8: Extract Date and Time into separate columns
df['Date'] = df['Date Time Served'].dt.date
df['Time'] = df['Date Time Served'].dt.time
print("\n'Date' and 'Time' columns extracted:")
print(df[['Date Time Served', 'Date', 'Time']].head())
```

```python
# Step 9: Calculate daily consumption
daily_consumption = df.groupby('Date')['Consumed (ml)'].sum().reset_index()
print("\nDaily consumption (ml):")
print(daily_consumption.head())
```

```python
# Step 10: Calculate total consumption by 'Alcohol Type'
consumption_by_alcohol_type = df.groupby('Alcohol Type')['Consumed (ml)'].sum().reset_index()
print("\nTotal consumption (ml) by Alcohol Type:")
print(consumption_by_alcohol_type.sort_values(by='Consumed (ml)', ascending=False))
```

```python
# Step 11: Calculate total consumption by 'Brand Name'
consumption_by_brand_name = df.groupby('Brand Name')['Consumed (ml)'].sum().reset_index()
print("\nTotal consumption (ml) by Brand Name:")
print(consumption_by_brand_name.sort_values(by='Consumed (ml)', ascending=False))
```

```python
# Ensure 'Date' column is in datetime format for plotting
daily_consumption['Date'] = pd.to_datetime(daily_consumption['Date'])
```

```python
# Step 12: Visualize Daily Consumption
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Consumed (ml)', data=daily_consumption)
plt.title('Daily Consumption Trend')
plt.xlabel('Date')
plt.ylabel('Consumed (ml)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 13: Visualize Consumption by Alcohol Type
plt.figure(figsize=(10, 6))
sns.barplot(x='Consumed (ml)', y='Alcohol Type', data=consumption_by_alcohol_type.sort_values(by='Consumed (ml)', ascending=False))
plt.title('Total Consumption by Alcohol Type')
plt.xlabel('Consumed (ml)')
plt.ylabel('Alcohol Type')
plt.tight_layout()
plt.show()
```

```python
# Step 14: Visualize Consumption by Brand Name (Top 10)
top_10_brands = consumption_by_brand_name.sort_values(by='Consumed (ml)', ascending=False).head(10)
plt.figure(figsize=(12, 7))
sns.barplot(x='Consumed (ml)', y='Brand Name', data=top_10_brands)
plt.title('Top 10 Brands by Total Consumption')
plt.xlabel('Consumed (ml)')
plt.ylabel('Brand Name')
plt.tight_layout()
plt.show()
```

```python
# Step 15: Calculate total consumption by 'Bar Name' and visualize it
consumption_by_bar = df.groupby('Bar Name')['Consumed (ml)'].sum().reset_index()
print("\nTotal consumption (ml) by Bar Name:")
print(consumption_by_bar.sort_values(by='Consumed (ml)', ascending=False))

plt.figure(figsize=(10, 6))
sns.barplot(x='Consumed (ml)', y='Bar Name', data=consumption_by_bar.sort_values(by='Consumed (ml)', ascending=False))
plt.title('Total Consumption by Bar Name')
plt.xlabel('Consumed (ml)')
plt.ylabel('Bar Name')
plt.tight_layout()
plt.show()
```

```python
# Step 16: Analyze consumption by time of day
# Define a function to categorize time into parts of the day
def get_time_of_day(time_obj):
    hour = time_obj.hour
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['Time of Day'] = df['Time'].apply(get_time_of_day)

consumption_by_time_of_day = df.groupby('Time of Day')['Consumed (ml)'].sum().reset_index()
# Define a specific order for the time of day categories for better visualization
time_of_day_order = ['Morning', 'Afternoon', 'Evening', 'Night']
consumption_by_time_of_day['Time of Day'] = pd.Categorical(consumption_by_time_of_day['Time of Day'], categories=time_of_day_order, ordered=True)
consumption_by_time_of_day = consumption_by_time_of_day.sort_values('Time of Day')

print("\nTotal consumption (ml) by Time of Day:")
print(consumption_by_time_of_day)

plt.figure(figsize=(8, 5))
sns.barplot(x='Time of Day', y='Consumed (ml)', data=consumption_by_time_of_day)
plt.title('Total Consumption by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Consumed (ml)')
plt.tight_layout()
plt.show()
```

```python
# Step 17: Summarize key insights
print("\n--- Project Summary and Key Insights ---")
print("1. Data Overview: The dataset contains consumption records across various bars, alcohol types, and brands, including opening balance, purchase, consumed, and closing balance in ml.")
print(f"2. No Missing Values: All columns have complete data, indicating a clean dataset for analysis.")
print(f"3. Top Alcohol Types by Consumption: {consumption_by_alcohol_type.sort_values(by='Consumed (ml)', ascending=False).iloc[0]['Alcohol Type']} was the most consumed alcohol type, followed by {consumption_by_alcohol_type.sort_values(by='Consumed (ml)', ascending=False).iloc[1]['Alcohol Type']}.")
print(f"4. Top Brand by Consumption: {consumption_by_brand_name.sort_values(by='Consumed (ml)', ascending=False).iloc[0]['Brand Name']} was the most consumed brand.")
print(f"5. Bar Performance: {consumption_by_bar.sort_values(by='Consumed (ml)', ascending=False).iloc[0]['Bar Name']} had the highest total consumption.")
print(f"6. Peak Consumption Times: Consumption is highest during the {consumption_by_time_of_day.sort_values(by='Consumed (ml)', ascending=False).iloc[0]['Time of Day']} hours, indicating potential peak operational periods for bars.")
print("\n--- End of Project ---")
```
